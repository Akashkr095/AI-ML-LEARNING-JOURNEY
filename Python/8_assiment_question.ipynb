```python
def print_digit(n):
    digit = []
    while n>=1:
        lastdigit = n%10
        digit.append(lastdigit)
        n = n//10
    for i in reversed(digit):
        print(i)
    
print_digit(312) 
```

```python
def cal(a,b,operation):
    if operation == "+":
        print("sum of a and b is = ",a+b)
    elif operation == "-":
        print("difference of a and b is = ",a-b)
    elif operation == "*":
        print("multiplication of a and b is = ",a*b)
    elif operation == "/":
        print("Division of a and b is = ", a/b)
    else:
        print("invaild operation")

a = float(input("Enter the 1st Num: "))
print("a = ",a)
b = float(input("Enter the 2nd Num: "))
print("b = ",b)
operation = input("Enter the 1st Num: ")
print("o = ",operation)

cal(a,b,operation)
```

```python
def isprime(n):
    if n <= 1:  
        return False

    for i in range(2,n):
        if n % i == 0:
            return False
        
    return True
isprime(11)
```
