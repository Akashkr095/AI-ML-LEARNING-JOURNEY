```python
def isprime(n):
    if n <= 1:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
    else:
        return True
    
isprime(2)
isprime(10)
isprime(13)
```

```python
list = [1,1,2,5,8,2,3,4,5]

seen = set()
dup = set()

for i in list:
    if i in seen:
        dup.add(i)
    else:
        seen.add(i)

print("Duplicate elements are:", dup)
```
