```python
# single inheritance

class Parent:
    def func1(self):
        print("This is parent class function")
    
class Child(Parent):
    pass

c1 = Child()
c1.func1()  # accessing parent class function from child class object
```

```python
# multi-level inheritance

class Employee:                  # parent class
    start_time = "9 AM"
    end_time = "5 PM"

class Administor(Employee):      # child class
    def __init__ (self,role):
        self.role = role

class Accountent(Administor):     # grand child class
    def __init__ (self,role,salary):
        super().__init__(role)
        self.salary = salary

a1 = Accountent("Manager",50_000)
print(a1.role, a1.salary, a1.start_time, a1.end_time)
```

```python
## multiple inheritance

class Teacher:
    def __init__ (self, salary):
        self.salary = salary


class Student:
    def __init__ (self, gpa):
        self.gpa = gpa  


class TA(Teacher , Student):
    def __init__ (self, salary , gpa , name,):
        Teacher.__init__(self, salary)
        Student.__init__(self, gpa)
        self.name = name


ta1 = TA(30_000 , 7.8 , "RAM")

print(ta1.name , ta1.gpa , ta1.salary)
```
