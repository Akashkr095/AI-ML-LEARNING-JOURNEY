# OOP


```python
class Student:  # class 
    subject = "Maths"  # class variable
    collage = "XYZ"
    year = 2019

st1 = Student()   # object
st2 = Student()

print(st1.subject)     # accessing class variable using object
print(st2.collage)
print(st1.year)

```

# Constructor


```python
class Student:
    def __init__ (self): # constructor
        print("Constructor called")

st1 = Student()   # object
st2 = Student()
st3 = Student()
```

```python
class Student:
    def __init__ (self,name): # constructor
        self.name=name
        print("Constructor called")

st1 = Student("Akash")   # object
st2 = Student("Amar")
st3 = Student("Anil")

print(st1.name)
print(st2.name)
```

```python
class Student:

    def __init__ (self,name,age,gender,cgpa): # constructor
        self.name=name     # instance variable or attribute
        self.age=age
        self.gender=gender
        self.cgpa=cgpa
        

    def disply(self):
        return self.cgpa # accessing instance variable using method or function

st1 = Student("Akash",20,"Male",8.5)   # object
st2 = Student("Amar",22,"Male",7.8)
st3 = Student("Anil",21,"Male",9.0)

print(st1.name,st1.age,st1.gender,st1.cgpa)
print(f'{st3.name} has a cgpa of : {st3.disply()}')
```

```python
class Dog:
    def __init__ (self,name,age): # parameterized constructor
        self.name = name          # attribute
        self.age = age            # attribute


    def bark(self):               # method
        print(f'I am {self.name}, {self.age} year old.')  # accessing attribute
        print(f'{self.name} say woof.  ')                 # accessing attribute

dog1 = Dog("Lily" , 3)            # object
dog2 = Dog("Sam", 2)              # object


print(dog1.name)                 # accessing attribute       

dog1.bark()                      # calling method
dog2.bark()                      # calling method
```
