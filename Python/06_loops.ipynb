# Break

```python
i = 1
while i <= 10:
    if i % 5 == 0:
        break
    print(i)
    i += 1
```

# Continue

```python
i = 1
while i <= 10:
    if i % 3 == 0:
        i+= 1
        continue  # remove multiple of 3
    print(i)
    i += 1
```

```python
# write odd num for 1 to 10
i = 1
while i <=10:
    if  i % 2 == 0:
        i +=1
        continue
    print(i)
    i +=1
```

# FOR LOOPS

```python
for i in range(11):
    print(i)
```

```python
# in is a membership operator

word = "hello"
for ch in word:
    print(ch)
```

```python
word = "hello"
if "o" in word:
    print(" o is present")
```

```python
word = "artificial intelligence"

count = 0

for ch in word:
    if ch=='i':
        count += 1
        
print("number of i:",count)
```

```python
# vowel cheak
word = "artificial"
count = 0

for ch in word:
    if ch == "a" or ch == "e" or ch== "i" or ch == "o" or ch == "j":
        count += 1
print("number of vowels in this word: ", count)

```

```python
# sum of n nqatural numbers

n = int(input("Enter the number"))

sum = 0

for i in range(n+1):
    sum += i
print(f'sum of {n} natural number:',sum)
```
