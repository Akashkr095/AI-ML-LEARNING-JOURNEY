```python
## public member

class Bankaccount:
    def __init__(self, name,balance):
        self.name = name
        self.balance = balance

acc1 = Bankaccount("bro", 1000)

print(acc1.balance) # Accessing public member
```

```python
## protected member

class Bankaccount:
    def __init__(self, name,balance):
        self.name = name
        self._balance = balance            # protected member

acc1 = Bankaccount("bro", 1000)

print(acc1._balance) # it can be accessed but use  only within class and subclass
```

```python
## private member

class Bankaccount:
    def __init__(self, name,balance):
        self.name = name
        self.__balance = balance

acc1 = Bankaccount("bro", 1000)

print(acc1.__balance) # it give error as it is private member


# print(acc1._Bankaccount__balance) ----- not recommended to access private member this way
```

```python
## getter and setter for private member

class Bankaccount:
    def __init__(self, name,balance):
        self.name = name
        self.__balance = balance

    def get_balance(self):       # get methods
        return self.__balance
    
    def set_balance(self, new_balance):  # set methods
        self.__balance = new_balance

acc1 = Bankaccount("bro", 1000)

print(acc1.get_balance()) # Accessing private member via getter
acc1.set_balance(2000)   # Modifying private member via setter
```
