```python
# factorial

def fact(n):
    fac = 1
    for i in range(1,n+1):
        fac = fac * i
    return fac
print(fact(5))
        
```

```python
def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)
print(fact(5))
```

```python
# sum of largest of 3 numbers
def largest(a,b,c):
    if (a > b) and (a > c):
        print(f'{a} is largest')
    elif (b>a) and (b>c):
        print(f'{b} is largest')
    else:
        print(f'{c} is largest')

largest(1,2,3)
```
