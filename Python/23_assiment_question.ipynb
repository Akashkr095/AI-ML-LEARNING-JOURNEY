```python
class BankAccount:
    def __init__ (self,account_number, owner_name, balance):
        self.account_number = account_number
        self.owner_name = owner_name
        self.balance = balance

    def deposit(self,add):
        self.add = add
        self.balance = self.balance + self.add
        print(f'{self.add} is deposite in our account. Now you balance is {self.balance}\n')

    def withdraw(self,sub):
        self.sub = sub

        if self.sub > self.balance:
            print("insufficent money in your account.\n")
        else:
            self.balance = self.balance  - self.sub
            print(f'{self.sub} is withdraw in our account. Now your balance is {self.balance}\n')

    
    def cheak_balance(self):
        print(f'Balance in your account : {self.balance}\n')

c1 = BankAccount( 1 , "Rahul" , 0)
c1.deposit(500)
c1.deposit(500)

c1.withdraw(200)
c1.withdraw(800)
c1.withdraw(100)
c1.cheak_balance()
```

```python
class Book:
    
    def __init__ (self,title,author):
        self.title = title
        self.author = author
        self.list_of_review = []
        

    def add_review(self,review):
        self.review = review
        self.list_of_review.append(review)
       # print("your review is added ")
    
    def disply_reviews(self):
        print(f'{self.title} all reviews')
        count = 1
        for i in self.list_of_review:
            print(count,". ",i)
            count += 1
        print("Total Review : ", count - 1)

b1 = Book("Atomic Habit" , "James Clear" )

b1.add_review("Nice book")
b1.add_review("must read")
b1.add_review("imporatant")
b1.disply_reviews()

b2 = Book("DO IT Today" , "Darius Foroux")

b2.add_review("Good book")
b2.add_review("student must read")



b2.disply_reviews()



```

```python
class Student:
    def __init__ (self,name , roll_no , marks):
        self.__name = None
        self.__roll_no = None
        self.__marks = None


        self.set_name(name)
        self.set_roll_no(roll_no)
        self.set_marks(marks)

    def get_name(self):            # getname
        return self.__name
    
    def set_name(self,new_name):   # setname
        if new_name.strip():
            self.__name = new_name

        else:
            print("Name can't be empty") 


    def get_roll_no(self):         # get roll no
        return self.__roll_no
    
    def set_roll_no(self,new_roll_no):    # set roll no
        if new_roll_no >= 1 and new_roll_no <= 100:
            self.__roll_no = new_roll_no
        else:
            print("Roll NO range is Between 1 to 100")


    def get_marks(self):           # get marks
        return self.__marks
    
    def set_marks(self,new_marks):   # set marks
        if new_marks < 0 :
            print("Marks can't be Negative")
        else:
            self.__marks = new_marks


s1 = Student("Akash", 4 , 75)

print(s1.get_name())
print(s1.get_roll_no())
print(s1.get_marks())


s2 = Student("Rahul" , 102 , -33)
print(s2.get_name())
print(s2.get_roll_no())
print(s2.get_marks())

s3 = Student("" , 10 , 53)
print(s3.get_name())
print(s3.get_roll_no())
print(s3.get_marks())

```

```python
class Shape:
    def area(self):
        print("Area")

class Circle(Shape):
    def area(self):
        print("Area of Circle is : PI x R^2")


class Rectangle(Shape):
    def area(self):
        print("Area of Rectangle is : Length x Breath ")


class Triangle(Shape):
    def area(self):
        print("Area of Triangle is : 1/2 X Length x Breath")


c1 = Circle()
c1.area()

r1 = Rectangle()
r1.area()

t1 = Triangle()
t1.area()

```

```python
class Vehicle:
    def __init__ (self,brand,model):
        self.brand  = brand
        self.model = model

class Car(Vehicle):
    def __init__ (self,brand,model,seats):
        super().__init__ (brand , model)
        self.seats = seats

class Bike(Vehicle):
    def __init__ (self,brand,model,engine_cc):
        super().__init__ (brand , model)
        self.engine_cc = engine_cc


c1 = Car("Tata" ,"Nexon" , 4 )
print(c1.brand)
print(c1.model)
print(c1.seats)

b1 = Bike("TVS","Apache RTR","200cc")
print(b1.brand)
print(b1.model)
print(b1.engine_cc)
```

```python
from abc import ABC ,  abstractmethod

class Employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass

class Intern(Employee):
    def calculate_salary(self):
        return 10_000
    
class FullTimeEmployee(Employee):
    def calculate_salary(self):
        return 50_000
    
class ContractEmployee(Employee):
    def calculate_salary(self):
        return 30_000
    
intern = Intern()
Full_Time = FullTimeEmployee()
Contract_Employee = ContractEmployee()

print("Intern Salary: ", intern.calculate_salary())
print("FullTime Salary : ", Full_Time.calculate_salary())
print("Contract Employee Salary: ", Contract_Employee.calculate_salary())

```

```python
class Person:
    def __init__ (self , name , age=None , address=None ):
        self.name = name
        self.age=age
        self.address=address

    def display(self):
        print("Name: " , self.name)
        print("Age: " , self.age)
        print("Address: ", self.address)

p1 = Person("Akash")
p1.display()
print("\n")

p2 = Person("Vikash" , 22)
p2.display()
print("\n")

p3 = Person("Suraj" , 25 , "Patna")
p3.display()
```

```python
class Player:
    player_count = 0

    def __init__ (self , name , level):
        self.name = name
        self.level = level
        Player.player_count += 1

    @classmethod
    def count_player(cls):
        return Player.player_count

p1 = Player("Akash" , "1st-Level")
p2 = Player("Ankit" , "2nd-Level")
p3 = Player("Tushar" , "3rd-Level")

print("TOTAL NUMBER OG PLAYER : ",Player.count_player())

```

```python
class Herbivore:
    def __init__ (self):
        self.plant_food = "Gress and Fruit"

    def eat_plant(self):
        print("Eating Plant -- " , self.plant_food)

class Carnivore:
    def __init__ (self):
        self.meat_food = "Meat and Fish"
    
    def eat_meat(self):
        print("Eating Meat --", self.meat_food)

class Omnivore:
    def __init__ (self):
        self.both_food = "Plant and Meat"

    def eat_both(self):
        print("Eating both Plant and Meat --",self.both_food)

class Bear(Herbivore , Carnivore , Omnivore):
    def __init__ (self):
        Herbivore.__init__(self)
        Carnivore.__init__(self)
        Omnivore.__init__(self)

    def nature(self):
        print("Bear is Omnivore Animals")

b = Bear()
b.eat_plant()
b.eat_meat()
b.eat_both()
b.nature()
```
