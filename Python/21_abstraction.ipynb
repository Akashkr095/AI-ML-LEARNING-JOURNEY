```python
from abc import ABC , abstractmethod

class Animals(ABC):
    
    @abstractmethod
    def speak(self):
        pass

class Lion(Animals):                 # class 1
    def speak(self):
        print("Lion say Roar")

class Cat(Animals):                  # class 2
    def speak(self):
        print("Cat say Meaoo")

class Dog(Animals):                   # class 3
    def speak(self):
        print("Dog say Bark")


# object

l = Lion()
l.speak()

c = Cat()
c.speak()

d = Dog()
d.speak()
```
