```python
while True:
    n = input("Enter the number (or type 'quit'): ")

    if n.lower() == "quit":
        break

    n = int(n)

    if n > 0:
        print("Positive")
    elif n < 0:
        print("Negative")
    else:
        print("Zero")
```
