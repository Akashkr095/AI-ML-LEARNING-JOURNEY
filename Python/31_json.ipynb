```python
import json

json_data = '{"name"  : " akash" , "age" : "20"}'
print(json_data)
print(type(json_data))
```

```python
jason_data= '{"name" : "Akash" , "City": "Patna"}'

pyt_obj = json.loads(jason_data)

print(pyt_obj)
print(type(pyt_obj))
```

```python
data = {
    "Fruit" : "Mango",
    "Vegetable" : "Tamato",
    "items" : [1,2]
}

json_string = json.dumps(data)

print(json_string)
print(type(json_string))

```
