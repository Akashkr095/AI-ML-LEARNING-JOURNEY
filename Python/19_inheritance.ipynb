```python
class Employee:
    start_time = "9 AM"
    end_time = "5 PM"


class Teacher(Employee):
    def __init__ (self, subject):
        self.subject = subject

t1 = Teacher("Math")

print(t1.subject, t1.start_time , t1.end_time)  
```

```python
class Employee:
    start_time = "9 AM"
    end_time = "5 PM"


    def change_time(self,new_end_time):
        self.end_time = new_end_time


class Teacher(Employee):
    def __init__ (self, subject):
        self.subject = subject

t1 = Teacher("Math")
t1.change_time("6 PM")


print(t1.subject, t1.start_time , t1.end_time)  
```

```python
class Employee:                     # parent class
    start_time = "9 AM"
    end_time = "5 PM"


class Teacher(Employee):            # child class
    def __init__ (self, subject):
        self.subject = subject

class Administor(Employee):         # child class
    def __init__ (self, role):
        self.role = role

t1 = Teacher("Math")

print(t1.subject, t1.start_time , t1.end_time)  

a1 = Administor("Manager")
print(a1.role, a1.start_time , a1.end_time)
```
