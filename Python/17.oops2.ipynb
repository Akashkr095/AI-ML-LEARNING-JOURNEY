# Attributes and Methods

attributes

```python
## class attribute

class Student:
    collage = "Abc Collage"  # class attribute


st1 = Student()

print(st1.collage)       # accessing class attribute using object
print(Student.collage)   # accessing class attribute using class name
```

```python
## instance attribute

class Student:
    def __init__ (self,name,cgpa):
        self.name = name
        self.cgpa = cgpa
    
st1 = Student("Akash",8.5)
print(st1.name)
print(st1.cgpa)

#print(Student.cgpa)     # accessing instance attribute using class name will give error
```

methods

```python
# instance method


class Laptop:
    storge_type = "SSD"  # class attribute

    def __init__ (self,brand,ram,stroge):  # constructor
        self.brand = brand                 # instance attribute
        self.ram = ram
        self.stroge = stroge

    def info(self):                                            # instance method
        print(f'Brand of Laptop is {self.brand}')              # accessing instance attribute inside  instance method
        print(f'RAM of Laptop is {self.ram}')
        print(f'Storage of Laptop is {self.stroge}')

        print(f'Storage type of Laptop is {self.storge_type}')  # accessing class attribute inside method



l1 = Laptop("Dell", "16gb" , "512gb")
l2 = Laptop("HP", "86gb" , "256gb")
l3 = Laptop("Lenavo", "16gb" , "512gb")

l1.info()
print()   
l2.info()
print()
l3.info()
```

```python
# static methods

class Laptop:
    storge_type = "SSD"  # class attribute

    def __init__ (self,brand,ram,stroge):  # constructor
        self.brand = brand                 # instance attribute
        self.ram = ram
        self.stroge = stroge

    def info(self):                                         # instance method
        print(f'Brand of Laptop is {self.brand}')             
        print(f'Storage of Laptop is {self.stroge}')

        print(f'Storage type of Laptop is {self.storge_type}')  


    @classmethod                                             # class method
    def class_info(cls):
        print(f'Storage type of Laptop is {cls.storge_type}')     


    @staticmethod                                          # static method
    def final_price(price,discount):
        f_price = price - (discount/100)*price
        print(f'Final price after discount is {f_price}')

   

l1 = Laptop("Dell", "16gb" , "512gb")

l1.final_price(50_000,10)
```

```python
class Product:
    count = 0
    def __init__ (self,name,price):
        self.name = name
        self.price = price
        Product.count += 1

    def get_info(self):
        print(f'Product name is {self.name} and price is {self.price}')

    @classmethod
    def total_products(cls):
        print(f'Total products created: {cls.count}')


    @staticmethod
    def discount_price(price,discount):
        f_price = price - (discount/100)*price
        print(f'Final price after discount is {f_price}')

p1 = Product("Laptop",50_000)
p2 = Product("Mobile",20_000)
p3 = Product("Tablet",15_000)
p4 = Product("Monitor",12_000)

p1.get_info()
p1.discount_price(p1.price,10)
```
